```python
# information of the varibles 
def clu_data(data):
    return pd.DataFrame({"Data Type":data.dtypes,"No of Unique Data":data.apply(lambda x: x.nunique(),axis=0),
                        "Levels":data.apply(lambda x: str(x.unique()),axis = 0),"Null_values":data.isnull().sum(),'null%':((data.isnull().sum().values/len(data))*100).round(2)})
```

```python
#optimum number of clusters using elbow curve
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(clu) 
    
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 8))
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 123)

kmeans.fit(new_clu)
```

```python
identified_clusters = kmeans.fit_predict(new_clu)
identified_clusters
```

```python
from kmodes.kprototypes import KPrototypes
from sklearn.metrics import davies_bouldin_score
```

```python
labels = kmeans.labels_
davies_bouldin_score(new_clu, labels)
```
